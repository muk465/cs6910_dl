```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
%matplotlib inline

ds = pd.read_csv('fashion-mnist_train.csv')
#print (ds.shape)
#convert to numpy array
data = ds.to_numpy()
#print (data.shape)
#print(ds)
#print(data)

label_train = data[:, 0]
pixel_train = data[:, 1:]
#print(label_train.shape,pixel_train.shape)

#using a dict with list 
idx={i:[] for i in range(10)}
#combining common label elements under keys
for j in range(len(label_train)):
    idx[label_train[j]].append(j)

for i in idx.keys():
    #print(i)
    rnd=random.choice(idx[i])
    im=pixel_train[rnd].reshape((28,28))
    plt.figure(i,figsize=(2,2))
    plt.imshow(im,cmap='gray')
```
